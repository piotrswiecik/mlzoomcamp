## Data prep takeaways

- totalvisits won't matter much - zero out the nulls
- pageviews won't matter much - same as above
- asymmetrique_activity_score - this is problematic, about 50% data missing
- asymmetrique_profile_score - same as above

Probably the best way is to compare 2 approaches - fillna with mean/mode versus drop the column. Note that columns are paired e.g. asymmetrique_activity_score and asymmetrique_activity_index.

```python
# take care of the blanks

df.totalvisits = df.totalvisits.fillna(0)
df.page_views_per_visit = df.page_views_per_visit.fillna(0)
```

```python
# take care of activity indices

# in this version we will drop the columns
df_alternate_nulls = df.copy()
df_alternate_nulls = df_alternate_nulls.drop(
    [
        "asymmetrique_activity_score",
        "asymmetrique_activity_index",
        "asymmetrique_profile_index",
        "asymmetrique_profile_score"],
        axis=1
    )

# in base version - use means/modes
df.asymmetrique_activity_score = df.asymmetrique_activity_score.fillna(df.asymmetrique_activity_score.mean())
df.asymmetrique_activity_index = df.asymmetrique_activity_index.fillna(df.asymmetrique_activity_index.mode()[0])
df.asymmetrique_profile_score = df.asymmetrique_profile_score.fillna(df.asymmetrique_profile_score.mean())
df.asymmetrique_profile_index = df.asymmetrique_profile_index.fillna(df.asymmetrique_profile_index.mode()[0])
```
